```python
import cv2
import matplotlib.pyplot as plt
```

```python
color_image = cv2.imread('/Users/mirjalol/Downloads/ZiClJf-1920w.jpg')
```

```python
(b, g, r) = cv2.split(color_image)
```

```python
equalized_b = cv2.equalizeHist(b)
equalized_g = cv2.equalizeHist(g)
equalized_r = cv2.equalizeHist(r)
```

```python
equalized_color_image = cv2.merge([equalized_b,
equalized_g, equalized_r])
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(color_image,
cv2.COLOR_BGR2RGB))
plt.title('Original Color Image')
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(equalized_color_image,
cv2.COLOR_BGR2RGB))
plt.title('Equalized Color Image (Channels Separately)')
plt.axis('off')
```

```python
plt.show()
```

```python
import cv2
import numpy as np
from skimage import exposure
import matplotlib.pyplot as plt
```

```python
image_source = cv2.imread('/Users/mirjalol/Downloads/ZiClJf-1920w.jpg',
cv2.IMREAD_GRAYSCALE)
image_target = cv2.imread('/Users/mirjalol/Downloads/single-fresh-red-strawberry-on-table-green-background-food-fruit-sweet-macro-juicy-plant-image-photo.jpg',
cv2.IMREAD_GRAYSCALE)
```

```python
matched_image = exposure.match_histograms(image_source,image_target)
plt.figure(figsize=(15,5))
```

```python
plt.subplot(3, 3, 1)
plt.imshow(image_source, cmap='gray')
plt.title('Source Image')
plt.axis('off')
```

```python
plt.subplot(3, 3, 2)
plt.hist(image_source.ravel(), 256, [0,256])
plt.title('Histogram (Source)')
```

```python
plt.subplot(3, 3, 4)
plt.imshow(image_target, cmap='gray')
plt.title('Target Image')
plt.axis('off')
```

```python
plt.subplot(3, 3, 5)
plt.hist(image_target.ravel(), 256, [0,256])
plt.title('Histogram (Target)')
```

```python
plt.subplot(3, 3, 7)
plt.imshow(matched_image, cmap='gray')
plt.title('Matched Image')
plt.axis('off')
```

```python
plt.subplot(3, 3, 8)
plt.hist(matched_image.ravel(), 256, [0,256])
plt.title('Histogram (Matched)')
plt.show()
```
