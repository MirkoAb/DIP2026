```
import cv2
```

```
from PIL import Image

# 1. Open the image
img_pil = Image.open("/content/image/media_1271a7f98c8f0a907643346bd952b678aaafc3a22.png")

# 2. Print image metadata
print(f"Type: {type(img_pil)}")
print(f"Size: {img_pil.size}")  # (Width, Height)
print(f"Mode: {img_pil.mode}")  # e.g., RGB

# 3. Display the image
img_pil.show()
```

```
import numpy as np

# Convert PIL image to NumPy array
img_np = np.array(img_pil)

# Print array shape and data type
print(f"Array Shape: {img_np.shape}") # (Height, Width, Channels)
print(f"Data Type: {img_np.dtype}")
```

```
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# 1. Read the image
img_plt = mpimg.imread("/content/image/media_1271a7f98c8f0a907643346bd952b678aaafc3a22.png")

# 2. Print shape
print(f"Shape: {img_plt.shape}")

# 3. Display it
plt.imshow(img_plt)
plt.axis('off') # Optional: hides the axes
plt.show()
```

```
from skimage import io, img_as_float
import matplotlib.pyplot as plt

# 1. Read the image
img_sk = io.imread("/content/image/media_1271a7f98c8f0a907643346bd952b678aaafc3a22.png")

# 2. Convert to float
img_float = img_as_float(img_sk)

# 3. Print min/max pixel values
print(f"Min value: {img_float.min()}")
print(f"Max value: {img_float.max()}")

# 4. Display
io.imshow(img_float)
plt.show()
```

```
edges = cv2.Canny(img_sk, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')
plt.show()
```
